```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
ggplot(mtcars, aes(x='wt',y='mpg', color='factor(gear)')) + geom_point()
```

```python
ggplot(mtcars, aes(x='wt',y='mpg', color='factor(gear)')) + geom_point() \
+facet_wrap('~gear',scales="free")
```

```python
(ggplot(mtcars, aes(x='wt',y='mpg', color='factor(gear)')) + geom_point() \
+facet_wrap('~gear',scales="free")).save('.//plotnine.1.pdf')
```
